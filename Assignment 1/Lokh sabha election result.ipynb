```python
import pandas as pd
import numpy as np
from scipy import stats
```

```python
results = pd.read_csv('2019_Results.csv')
```

```python
results
```

```python
#extract Karnataka votes
karnatakavotes = results[results["State"] == "Karnataka"]
#Print Karnataka votes
karnatakavotes
```

```python
Totalvotes = karnatakavotes["Total Votes"]
#Measure of Central Tendency
#Mean
print("Mean of total votes of Karnataka is :") 
np.mean(Totalvotes)
```

```python
#Median
print("Median of total votes of Karnataka is :") 
np.median(Totalvotes)
```

```python
#Mode
print("Mode of total votes of Karnataka is :") 
stats.mode(Totalvotes)
```

```python
#Standard Deviation
print("SD of total votes of Karnataka is :") 
np.std(Totalvotes)
```

```python
#Max
print("Max no total votes of Karnataka is :") 
np.max(Totalvotes)

```

```python
#Min
print("Min no total votes of Karnataka is :") 
np.min(Totalvotes)
```

```python
#Skewness
print("Skewness total votes of Karnataka is :") 
stats.skew(Totalvotes)
```

```python
#Bagalkot votes
bvotes=karnatakavotes[karnatakavotes["Constituency"]=="Bagalkot"]
bvotes
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
# Line graph for total votes across constituencies in Bagalkot
btotalvotes = bvotes['Total Votes']
btotalvotes.plot()
plt.title('Total votes accross Constituencies')
plt.ylabel('Total Votes')
plt.xlabel('Constituency serial number')
plt.rcParams['figure.figsize'] = (20.0, 10.0)
```

```python
# Number of residential property transactions in Aug 2018 by property type
Partyvotes = bvotes["Party"].value_counts()
Partyvotes.sort_values().plot(kind='barh')
plt.title('Party count ')
plt.ylabel('Party Name')
plt.xlabel('Count')
```

```python
#Pie Chart
sale_volume = pd.DataFrame(bvotes['Party'].value_counts())
sale_volume.columns=['Count']
sale_volume.plot(kind='pie', y='Count')
```

```python
# Price distribution box plot
bvotes['Total Votes'].plot(kind="box")
plt.title('House Price distribution by Property Type')
plt.ylabel('No of votes')
#Outliers exist
```
